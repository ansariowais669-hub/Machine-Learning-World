What is an Outlier ?

An outlier is a data point that is distant from all other observations. A data point that lies outside the overall distribution fo the dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

What are the criterias to identify an outlier ?

1. Data point that falls 1.5 times outside of an interquartile range above the third quartile and below the first quartile.
2. Data point that falls outside 3 standard deviations. We can use a Z-score and if the Z-score falls outside of 2 standard deviation.

What is the reason for an outlier to exist in a dataset ?

1. Variability in Data.
2. An experimental measurement error

What are the impacts of having outliers in a dataset ?

1. It causes various problems during our statistical data.
2. It may cause a significant impact on the mean and the standard deviation.

Various ways of finding the outlier.

1. Using the scatter plot
2. Box plot
3. Using Z-score
4. Using the Interquartile range

```python
dataset= [11,10,12,14,12,15,14,13,15,102,12,14,17,19,107, 10,13,12,14,12,108,12,11,14,13,15,10,15,12,10,14,13,15,10]
```

## Detecting outlier using Z-score

# Using Z-score

Formula for Z-score = (Observation - Mean)/Standard Deviation

Z= (X-u)/sigma

```python
outliers = []
def detect_outliers(data):

    threshold = 3 #within third standard deviation normal points
    mean = np.mean(data)
    std = np.std(data)

    for i in data:
        z_score = (i - mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
outlier_pt = detect_outliers(dataset)
outlier_pt
```

## InterQuantile Range

75%-25% values in the dataset

# Steps

1. Arrange the data in increasing order
2. Calculate first(q1) and third quartile(q3)
3. Find the Interquantile ramge (q3 - q1)
4. Find the lower bound q1x1.5 and the upper bound q3x1.5

   Anything that lies outside of lower and upper bound is an outlier

```python
## Perform all the steps of IQR
sorted(dataset)
```

```python
quantile1 , quantile3 = np.percentile(dataset,[25,75])
print(quantile1 , quantile3)
```

```python
# Find the IQR

iqr_value = quantile3 - quantile1
print(iqr_value)
```

```python
# Find the lower bound value and upper bound value

lower_bound_val = quantile1 - (1.5 * iqr_value)
upper_bound_val = quantile3 + (1.5 * iqr_value)

print(lower_bound_val , upper_bound_val)
```

```python
outlier = []
for i in dataset:
    if(i > upper_bound_val or i < lower_bound_val):
        outlier.append(i)

print(outlier)
```
