# Working with Dates

```python
# Converting to date time datatype

date['date'] = pd.to_datetime(date['date'])
date.info()
```

```python
date['date_year'] = date['date'].dt.year
date['date_month_no'] = date['date'].dt.month
date['date_month_name'] = date['date'].dt.month_name()
date['date_day'] = date['date'].dt.day # Day number of month
date['date_dayOfWeek'] = date['date'].dt.dayofweek # Day of week
date['date_dow_name'] = date['date'].dt.day_name() # day by name

date.head()
```

```python
# Is weekend ?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Saturday','Sunday']),1,0)
date.drop(columns = ['product_id','city_id','orders']).head()
```

```python
# Extracting week of the year

date['date_week'] = date['date'].dt.isocalendar().week

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Extracting Quarter of the Year

date['quarter'] = date['date'].dt.quarter
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Extracting Semesters

date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Extracting Time elapsed between dates

import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

np.round(
    (today - date['date']) / np.timedelta64(1, 'D') / 30.44,
    0
) # 30.44 represents average days per month
```

```python
# Converting to datetime datatypes
time['date'] = pd.to_datetime(time['date'])
time.info()
```

```python
time['hour'] = time['date'].dt.hour
time['minute'] = time['date'].dt.minute
time['seconds'] = time['date'].dt.second

time.head()
```

```python
# Extracting only time part from date column

time['time'] = time['date'].dt.time

time.head()
```

```python
# Time difference

today - time['date']
```

```python
# in seconds

(today - time['date'])/np.timedelta64(1,'s')
```

```python
# in minutes

(today - time['date'])/np.timedelta64(1,'m')
```

```python
# in hours

(today - time['date'])/np.timedelta64(1,'h')
```
