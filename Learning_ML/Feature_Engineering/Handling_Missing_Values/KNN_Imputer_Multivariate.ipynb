```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.impute import KNNImputer , SimpleImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('titanic.csv',usecols=['Age','Fare','Pclass','Survived'])
df.head()
```

```python
df.isnull().mean()*100
```

```python
X = df.drop(columns='Survived')
y=df['Survived']

X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=2)
X_train.head()
```

```python
knn = KNNImputer() # weights='uniform' by default

X_train_trf = knn.fit_transform(X_train)
X_test_trf = knn.transform(X_test)
```

```python
pd.DataFrame(X_train_trf,columns=X_train.columns)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf , y_train)

y_pred = lr.predict(X_test_trf)

accuracy_score(y_pred,y_test)
```

```python
# Comparison with SimpleImputer --> Mean

si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf2 , y_train)

y_pred = lr.predict(X_test_trf2)

accuracy_score(y_pred,y_test)
```

```python
knn = KNNImputer(weights='distance')

X_train_trf3 = knn.fit_transform(X_train)
X_test_trf3 = knn.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf3 , y_train)

y_pred = lr.predict(X_test_trf3)

accuracy_score(y_pred,y_test)
```
