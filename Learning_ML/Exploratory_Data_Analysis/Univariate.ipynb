```python
import pandas as pd
import seaborn as sns
df=pd.read_csv('train.csv')
df.head()
```

## Categorical Data

```python
# a. Countplot
```

```python
sns.countplot(x = 'Survived', data = df)
```

```python
df['Survived'].value_counts()
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(x = 'Pclass', data = df)
```

```python
sns.countplot(x = 'Embarked' , data = df) #Embarked is the code of cities from where people got on Titanic
```

# Piechart

```python
df['Survived'].value_counts().plot(kind = 'pie',explode=[0.02,0],autopct='%.2f%%')
```

```python
df['Pclass'].value_counts().plot(kind = 'pie',explode=[0.02,0.02,0.02],autopct='%.2f%%')
```

```python
df['Embarked'].value_counts().plot(kind = 'pie',explode=[0.02,0.02,0.02],autopct='%.2f%%')
```

```python
df['Sex'].value_counts().plot(kind = 'pie',explode=[0.02,0],autopct='%.2f%%')
```
