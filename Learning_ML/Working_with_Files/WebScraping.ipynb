```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/137.0.0.0 Safari/537.36"
    )
}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
# webpage contains the whole HTML of that page
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
for i in soup.find_all('div',style='height:auto;'):
    print(i.text.strip())
```

```python
div = soup.find_all("div", class_="companyCardWrapper__tertiaryInformation")
for divs in div:
    first_a = divs.find("a")
    first_span = first_a.find("span")
    print(first_span.text.strip())
```

# Or we can simply go through the whole div and get content each time

```python
name=[]
rating=[]
div = soup.find_all('div',class_='companyCardWrapper')

for i in div:
    name.append(i.find('h2',class_='companyCardWrapper__companyName').text.strip())
    a = i.find('a',class_='companyCardWrapper__ActionWrapper')
    rating.append(a.find('span',class_='companyCardWrapper__ActionCount').text.strip())
```

```python
d = {'name':name, 'rating':rating}
df = pd.DataFrame(d)
df
```

# And now to access all the pages of the website

```python
final = pd.DataFrame()

for j in range(1,3):
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/137.0.0.0 Safari/537.36"
        )        
    }
    webpage = requests.get(url,headers=headers).text
    soup = BeautifulSoup(webpage,'lxml')
    name=[]
    rating=[]
    div = soup.find_all('div',class_='companyCardWrapper')

    for i in div:
        name.append(i.find('h2',class_='companyCardWrapper__companyName').text.strip())
        a = i.find('a',class_='companyCardWrapper__ActionWrapper')
        rating.append(a.find('span',class_='companyCardWrapper__ActionCount').text.strip())
        d = {'name':name, 'rating':rating}
        
        df = pd.DataFrame(d)

    final = pd.concat([final, df], ignore_index=True)


final
```
